```python
from skimage.metrics import structural_similarity
%matplotlib inline
for ik in fnamelist:
    print(ik)
    img = cv2.imread(ik)
    
    #plt.figure(figsize = (10,10))
    print("original image grayscale")
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plt.imshow(img_gray, cmap='gray',aspect='auto')
    plt.axis('off')
    plt.show()
    
    
    


    methods = ['cv2.TM_CCOEFF']#, 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
                #'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
    
    img_cropn = img_gray
    imgc = img_cropn.copy()
    template = img_sig.copy()
    for meth in methods:
        wt, ht = template.shape[::-1]
        #img = img2.copy()
        method = eval(meth)
        # Apply template Matching
        res = cv2.matchTemplate(imgc,template,method)
        #print(res)
        
        print(np.max(res))
        threshold = 0.99*np.max(res)
        loc = np.where( res >= threshold)

        for pt in zip(*loc[::-1]):
            #print(pt)
            cv2.rectangle(imgc, pt, (pt[0] + wt, pt[1] + ht), (0,0,255), 2)
            #break


        plt.subplot(121)
        plt.imshow(res,cmap = 'gray')
        plt.title('Matching Result')
        plt.xticks([])
        plt.yticks([])
        plt.subplot(122)
        plt.imshow(imgc)#,cmap = 'gray')
        plt.title('Detected Point')
        plt.xticks([])
        plt.yticks([])
        plt.suptitle(meth)
        plt.show()
        
        plt.subplot(121)
        plt.title("image inside the detected box")
        fil_img = img_cropn.copy()[pt[1]:pt[1] + ht, pt[0]:pt[0] + wt]
        plt.imshow(fil_img)
        #plt.show()
        plt.subplot(122)
        plt.title("flipped image inside the detected box")
        fil_img_flip = np.flip(fil_img)
        plt.imshow(fil_img_flip)
        plt.show()
        
        plt.subplot(121)
        (score, diff) = structural_similarity(template, fil_img,full=True)
        plt.title("Image similarity same :"+ str(score))
        plt.imshow(diff)
        #plt.show()
        plt.subplot(122)
        (score, diff) = structural_similarity(template, fil_img_flip,full=True)
        plt.title("Image similarity flipped :"+ str(score))
        plt.imshow(diff)
        plt.show()
```
